```python
import numpy as np
import matplotlib.pyplot as plot
```

 # Get x values of the sine wave

```python
time = np.arange(0, 10, 0.1);
print(time)
```

```python
list1 = time.tolist()
print(list1)
```

# Amplitude of the sine wave is sine of a variable like time

```python
amplitude = np.sin(time)
print(amplitude)
list2 = amplitude.tolist()
print(list2)
```

# Plot a sine wave using time and amplitude obtained for the sine wave

```python
plot.plot(time, amplitude)

# Give a title for the sine wave plot
plot.title('Sine Wave')

# Give x axis label for the sine wave plot
plot.xlabel('Time')

# Give y axis label for the sine wave plot
plot.ylabel('Amplitude = sin(Time)')

plot.grid(True, which = 'both')
plot.axhline(y = 0, color = 'r')

# Display the sine wave
plot.show()
```

```python
time_1 = np.arange(0, 1, 0.001);
amplitude_1 = np.cos(2*np.pi*3*time_1)

plot.plot(time_1, amplitude_1)

# Give a title for the cos wave plot
plot.title('Cosine Wave')

# Give x axis label for the cos wave plot
plot.xlabel('Time')

# Give y axis label for the cos wave plot
plot.ylabel('Amplitude = cos(Time)')

plot.grid(True, which = 'both')
plot.axhline(y = 0, color = 'r')

# Display the sine wave
plot.show()
```
